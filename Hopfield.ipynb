```python
# makes a random permutation of output units
def get_random_permutation():
    res = np.random.permutation(63)
#     res = res+1  #so that numbers are from 1 to 63
    return res
```

```python
# checks if outputs from one iteration have converged
def has_converged(y_before, y_after):
    if np.all(y_before == y_after):
        return True
    return False
```

```python
def readable(X):
    X = ['#' if x==1 else x for x in X]
    X = ['o' if x==-1 else x for x in X]
    X = np.reshape(X,(9,7))
    return X
```

```python
# calculates W accumulatively using sample_count
# order of samples is a1, b1, c1, ...
# samples_count increases as more and more samples are trained by the net
def calculate_W(samples_count):
    W = np.zeros((63,63), dtype=np.int)
    for i in range(samples_count):
        sample = patterns[i]
        sample = np.reshape(sample, (63,1))
        W = np.matmul(sample, np.transpose(sample)) + W
    identity = np.identity(63) * samples_count
    W = W - identity
    return W
```

```python
def activation(y_in):
    theta = 0
    if y_in > theta:
        return 1
    elif y_in == theta:
        return y_in
    elif y_in < theta:
        return -1
    
```

```python
def summation(Y, W, index):
    W_col = W[:][index]
#     print(W_col)
#     print(Y)
    return np.matmul(Y, W_col)
```

```python
import random
import math
# returns a new version of input with mistakes
def make_mistake(percent, input):
    X = input.copy()
    X = np.reshape(X, (1,63))[0]
    num_mistake = math.ceil(percent*63)
    randoms = random.sample(range(63), num_mistake)
    for i in range(num_mistake):
        X[randoms[i]] = X[randoms[i]]*-1
    return np.reshape(X, (9,7))

# returns a new version of input with missings
def make_missing(percent, input):
    X = input.copy()
    X = np.reshape(X, (1,63))[0]
    num_missing = math.ceil(percent*63)
    print(num_missing)
    randoms = random.sample(range(63), num_missing)
    for i in range(num_missing):
        X[randoms[i]] = 0
    return np.reshape(X, (9,7))
```

```python
def train(training_patterns):
    samples_count = len(training_patterns)
    W = calculate_W(samples_count)
    print('W computed successfully')
    return W
  
def recall_one(pat, training_patterns):
    W = train(training_patterns)
    permutation = get_random_permutation()
   
    X = pat.copy()
    X = np.reshape(X, (1,63))[0]
    Y = X.copy()
    print('pattern is \n', readable(np.reshape(pat, (1,63))[0]))
    while_cnt = 0
    
    while True:
        while_cnt = while_cnt + 1
        for j in range(63):
            # updating permutation[j]
            index = permutation[j] 
            y_in = X[index] + summation(Y, W, index)
            y = activation(y_in)
            Y_new = Y.copy()
            Y_new[index] = y
            Y = Y_new.copy()
        if has_converged(Y, Y_new):
            print('y{} converged after {} iterations'.format(index, while_cnt))
            break # go out of while loop
            
    print('&&&&&&&&&&&&&&&&&&&&&&&&&')
    print(readable(Y_new))
    print('****************************')
    return
        
```

```python
t_arr = patterns[:2]
character = b1
test = [make_mistake(0.1, character), make_mistake(0.2, character), make_mistake(0.3, character),
       make_mistake(0.4, character), make_mistake(0.5, character), make_mistake(0.6, character),
       make_mistake(0.7, character), make_mistake(0.8, character), make_mistake(0.9, character)]
test_missing = [make_missing(0.1, character), make_missing(0.2, character), make_missing(0.3, character),
       make_missing(0.4, character), make_missing(0.5, character), make_missing(0.6, character),
       make_missing(0.7, character), make_missing(0.8, character), make_missing(0.9, character)]

for i in range(len(test)):
    print('{}0 percent error:'.format(i+1))
    recall_one(test_missing[i], t_arr)
```
